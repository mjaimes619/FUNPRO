````python
n=10
s=0
m=0
sueldo=[]
for e in range(n):
  sueldo.append(int(input(f"Ingrese el sueldo del trabajador {e+1}: ")))
  s=sueldo[e]+s
prom=s/n
print(f"El sueldo promedio de la empresa es {prom}: ")
for e in range(n):
  if sueldo[e]>prom: m=m+1
print(f"{m} trabajadores ganan mas que el promedio")
````

````python
n=5
notas=[]
t=0
for e in range (n):
  notas.append(int(input(f"Ingrese la nota numero {e+1}.: ")))
  t=notas[e]+t
vm = min(notas);
print(f"{vm} es la peor nota ")
notas.remove(vm)
tf=t-vm
prom=tf/(n-1)
print(f"El promedio de el estudiante sacando la peor nota es {prom}")
````

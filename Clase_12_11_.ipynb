```python
def contador():
  notas=[]
  cantidad=int(input("Ingrese la cantidad de notas: "))
  for i in range(cantidad):
    nota=float(input(f"dijite sus notas:{i+1} "))
    notas.append(nota)
    print(notas)
  c=0
  for a in notas:
    if a>=3:
      c+=1
      print(f"van {c} notas")
  print(f"{c} notas fueron mayores o iguales a 3")
  return cantidad,notas

def promedio(notas,cantidad):
  total=sum(notas)
  promedio=total/cantidad
  print(f"su promedio es: {promedio}")

a,b=contador()
promedio(b,a)
```

```python
def contador():

  nts=[]
  cnt=int(input("Ingrese la cantidad de notas: "))
  for i in range(cnt):
    nt=float(input(f"dijite sus notas:{i+1} "))
    nts.append(nt)
    print(nts)
  c=0
  for a in nts:
    if a>=3:
      c+=1
      print(f"van {c} notas")
  print(f"{c} notas fueron mayores o iguales a 3")
  return cnt,nts

def promedio(nts,cnt):
  ttl=sum(nts)
  prom=ttl/cnt
  print(f"su promedio es: {prom}")

a,b=contador()
promedio(b,a)
```

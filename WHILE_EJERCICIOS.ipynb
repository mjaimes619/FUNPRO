```python
c=int(input("Ingrese la cantidad de numeros que va a operar: "))
for e in range(c):
  n=int(input(f"Ingrese el numero {e+1}: "))
  cubo=n*n*n
  print(f"El cubo de {n} es {cubo}")
```

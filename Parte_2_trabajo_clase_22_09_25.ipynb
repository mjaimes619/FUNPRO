```python
## Calculadora de indice de masa corporal if/elif/else
peso=float(input("Ingrese su peso (kg): "))
altura=float(input("Ingrese su altura (m): "))
imc= peso/(altura*altura)

if imc<18.5:
  print(f"Con un IMC de {imc}, tiene un peso bajo")
elif imc>=18.5 and imc<25:
  print(f"Con un IMC de {imc}, tiene un peso normal")
elif imc>=25 and imc<30:
  print(f"Con un IMC de {imc}, se identifica sobrepeso")
else:
  print(f"Con un IMC de {imc}, se identifica obesidad")
```

```python
## Calculadora de indice de masa corporal if/else
peso=float(input("Ingrese su peso (kg): "))
altura=float(input("Ingrese su altura (m): "))
imc= peso/(altura*altura)

if imc<18.5:
  print(f"Con un IMC de {imc}, tiene un peso bajo")
else:
  if imc>=18.5 and imc<25:
    print(f"Con un IMC de {imc}, tiene un peso normal")
  else:
    if imc>=25 and imc<30:
      print(f"Con un IMC de {imc}, se identifica sobrepeso")
    else:
      print(f"Con un IMC de {imc}, se identifica obesidad")
```

No se puede hacer case/match debido a que peso y altura manejan rangos, esto hara que siempre imc de un resultado distinto e imprecedible

```python
## Clasificacion de triangulos if/elif/else
l1=float(input("Ingrese el valor del lado 1 del triangulo "))
l2=float(input("Ingrese el valor del lado 2 del triangulo "))
l3=float(input("Ingrese el valor del lado 3 del triangulo "))

if l1==l2 and l2==l3:
  print("Se tiene un triangulo Equilatero")
elif l1==l2 or l1==l3 or l2==l3:
  print("Se tiene un triangulo Isosceles")
else:
  print("Se tiene un triangulo Escaleno")
```

```python
## Verificar si un numero es positivo, negativo o cero con if/elif/else
n=int(input("Ingrese un numero cualquiera: "))
if n<0:
  print(f"El numero {n} es negativo")
elif n==0:
  print(f"El numero {n} es cero")
else:
  print(f"El numero {n} es positivo")
```

```python
## Verificar si un numero es positivo, negativo o cero con if/else
n=int(input("Ingrese un numero cualquiera: "))
if n<0:
  print(f"El numero {n} es negativo")
else:
  if n==0:
   print(f"El numero {n} es cero")
  else:
   print(f"El numero {n} es positivo")
```
